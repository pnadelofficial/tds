```python
def splitEmails(thread):
  if (thread.startswith('To:') or thread.startswith("ts \n\n\nTo:") or thread.startswith('s \n\n\nTo:') or thread.startswith('nts \n\n\nTo:') or thread.startswith('EPA FOIA 5/12/2016 \nFlint Water Documents \n\n\nTo:') or thread.startswith('Cc:') or thread.startswith('nts \n\n\nCc:') or thread.startswith('ater Documents \n\n\nTo:') or thread.startswith('er Documents \n\n\nTo:')):
    emails_from_thread = thread.split('From:')
    emails_from_thread[0] = emails_from_thread[0] + 'From:' + emails_from_thread[1]
    emails_from_thread.remove(thread.split('From:')[1])
    return emails_from_thread
  else:
    return None
```

```python
def findLine(raw_email, regex):
  line_search = re.search(regex, raw_email, re.DOTALL)
  if line_search:
    return line_search.group().strip()
  else:
    return None

def findNames(line):
  if type(line) == str:
    name_re = r"([A-Z][a-z]+, [A-Z][a-z]+)"
    name_search = re.findall(name_re, line)
    return name_search

def findDates(line):
  if type(line) == str:
    date_re = r"[A-Z][a-z]+\s\d\d,\s\d\d\d\d"
    date_search = re.findall(date_re, line)
    if len(date_search) == 0:
      date_re = r'\d\/\d\d\/\d\d\d\d' 
      date_search = re.findall(date_re, line)   
    return date_search

def searchOrgEmail(org_email): ## run on [0] in splitEmails list
  org_email = org_email.strip()
  to_line = findNames(findLine(org_email, "(?<=To: )(.*)(?=Cc:)"))
  cc_line = findNames(findLine(org_email, "(?<=Cc: )(.*)(?=From:)"))
  from_line = findNames(findLine(org_email, "(?<=From: )(.*)(?=Sent:)"))
  date_line = re.findall(r'[A-Z][a-z]+\s\d\d,\s\d\d\d\d|\d\/\d\d\/\d\d\d\d|\d\d\/\d\d\/\d\d\d\d|\d\/\d\/\d\d\d\d|[A-Z][a-z]+\s\d,\s\d\d\d\d',org_email)
  
  # findDates(findLine(org_email, "(?<=Sent: )(.*)(?=Sub)"))
  # if findLine(org_email, "(?<=Sent: )(.*)(\n\s)(?=Sub)") == None:
  #   date_line = findDates(findLine(org_email, "(?<=Date: )(.*)(?=Sub)"))
  # if date_line == []:
  #   date_line = re.findall(r"[A-Z][a-z]+\s\d\d,\s\d\d\d\d",org_email)
  # if date_line == []:
  #   date_line = re.findall(r"\d\/\d\d\/\d\d\d\d",org_email)
  if len(org_email.split('Sub')) > 1: 
    rest = org_email.split('Sub')[1]
  else:
    rest = org_email.split('Sub')[0]
  return to_line, cc_line, from_line, date_line, rest

def searchThreadEmail(thread_email): ## run on [1] - [n] in splitEmails list
  thread_email = thread_email.strip()
  to_line = findNames(findLine(thread_email, "(?<=To: )(.*)(?=Cc:)"))
  cc_line = findNames(findLine(thread_email, "(?<=Cc: )(.*)(\n\s)"))
  from_line = findNames(findLine(thread_email, "(?<=From:)(.*)(?=Sent:)"))
  date_line = re.findall(r'[A-Z][a-z]+\s\d\d,\s\d\d\d\d|\d\/\d\d\/\d\d\d\d|\d\d\/\d\d\/\d\d\d\d|\d\/\d\/\d\d\d\d|[A-Z][a-z]+\s\d,\s\d\d\d\d',thread_email)
  
  # findDates(findLine(thread_email, "(?<=Sent: )(.*)(?=Sub)"))
  # if findLine(thread_email, "(?<=Sent: )(.*)(\n\s)(?=Sub)") == None:
  #   date_line = findDates(findLine(thread_email, "(?<=Date: )(.*)(?=Sub)"))
  # if (date_line == []):
  #   date_line = re.findall(r"[A-Z][a-z]+\s\d\d,\s\d\d\d\d",thread_email)
  # if (date_line == []):
  #   date_line = re.findall(r"\d\/\d\d\/\d\d\d\d",thread_email)
  if len(thread_email.split('Sub')) > 1:
    rest = thread_email.split('Sub')[1]
  else:
    rest = thread_email.split('Sub')[0]
  return to_line, cc_line, from_line, date_line, rest

def formatParser(email):
  if (email.startswith('To:') or email.startswith("ts \n\n\nTo:") or email.startswith('s \n\n\nTo:') or email.startswith('nts \n\n\nTo:') or email.startswith('EPA FOIA 5/12/2016 \nFlint Water Documents \n\n\nTo:') or email.startswith('Cc:') or email.startswith('nts \n\n\nCc:') or email.startswith('ater Documents \n\n\nTo:') or email.startswith('er Documents \n\n\nTo:')):
    return searchOrgEmail(email)
  else:
    return searchThreadEmail(email) 
```

### Helper Functions for parsing 

```python
def extractFrom(raw_email):
  from_re = r"(.*?)(?=Sent:)"
  from_search = re.search(from_re, raw_email, re.DOTALL)
  if from_search:
    return from_search.group().strip()
  else:
    from_re2 = r"(.*?)(?=Date:)"
    from_search2 = re.search(from_re2, raw_email, re.DOTALL)
    if from_search2:
      return from_search2.group().strip()
    else:
      return None
```

```python
def extractDate(raw_email):
  sent_re = r"(?<=Sent:)(.*?)(?=To:)"
  sent_search = re.search(sent_re, raw_email, re.DOTALL)
  if sent_search:
    return sent_search.group().strip()
  else:
    date_re = r"(?<=Date:)(.*?)(?=To:)"
    date_search = re.search(date_re, raw_email, re.DOTALL)
    if date_search:
      return date_search.group().strip()
    else:
      return None
```

```python
def extractTo(raw_email):
  to_re = r"(?<=To:)(.*?)(?=Cc:)"
  to_search = re.search(to_re, raw_email, re.DOTALL)
  if to_search:
    return to_search.group().strip()
  else:
    to_subject = r"(?<=To:)(.*?)(?=Subject:)"
    to_search_subject = re.search(to_subject, raw_email, re.DOTALL)
    if to_search_subject:
      if "Subject:" in to_search_subject.group().strip():
        to_search_subject.group().strip().split('Subject:')
        return to_search_subject.group().strip()
      else:
        return to_search_subject.group().strip()
    else:
      return None
```

```python
def extractCc(raw_email):
  cc_re = r"(?<=Cc:)(.*?)(?=Subject:)"
  cc_search = re.search(cc_re, raw_email, re.DOTALL)
  if cc_search:
    return cc_search.group().strip()
  else:
    return None
```

```python
def extractSubject(raw_email):
  subject_re = r"(?<=Subject:)(\s*)[^\n]+"
  subject_search = re.search(subject_re, raw_email, re.DOTALL)
  if subject_search:
    return subject_search.group().strip()
  else:
    return None
```

```python
def extractBody(raw_email, subject):
  body_re = f"(?<=Subject:)(.*)"
  body_search =  re.search(body_re, raw_email, re.DOTALL)
  if body_search:
    return body_search.group().replace(subject, '').strip()
  else:
    return None
```

```python
def cleanEntry(entry):
  if '\n' in entry:
    entry = entry.split('\n')[0]
    entry = re.sub(r'(?<=\<)(.*)(?=\>)', '' ,entry).replace("<", '').replace(">", '').replace('"', '').replace('|', '').replace("'", '').strip()
    entry = re.sub(r'(?<=\()(.*)(?=\))', '', entry).replace("(", '').replace(")", '') ## @TODO: search for names in parentheses, that is regex = ([A-Z][a-z]+, [A-Z][a-z]+)  
    entry = re.sub(r'(?<=\[)(.*)(?=\])', '', entry).replace("[", '').replace("]", '')
    if (',' not in entry) and (len(entry.split(' ')) > 1):
      entry = f"{entry.split(' ')[1]}" + ', ' + f"{entry.split(' ')[0]}"
    if (len(entry.split(',')) > 1) and (len(entry.split()) > 2):
      entry = entry.split()[0] + entry.split()[1]
    if ('@' in entry):
      entry = re.sub(r'((.*)\s)','',entry)
    if (' ' not in entry) and (len(entry.split(',')) > 1):
      entry = entry.split(',')[0]+" "+entry.split(',')[1] 
    return entry
  else:
    entry = re.sub(r'(?<=\<)(.*)(?=\>)', '' ,entry).replace("<", '').replace(">", '').replace('"', '').replace('|', '').replace("'", '').strip()
    entry = re.sub(r'(?<=\()(.*)(?=\))', '', entry).replace("(", '').replace(")", '')
    entry = re.sub(r'(?<=\[)(.*)(?=\])', '', entry).replace("[", '').replace("]", '')
    if (',' not in entry) and (len(entry.split(' ')) > 1):
      entry = f"{entry.split(' ')[1]}" + ', ' + f"{entry.split(' ')[0]}"
    if (len(entry.split(',')) > 1) and (len(entry.split()) > 2):
      entry = entry.split()[0] + entry.split()[1]
    if ('@' in entry):
      entry = re.sub(r'((.*)\s)','',entry)
    if (' ' not in entry) and (len(entry.split(',')) > 1):
      entry = entry.split(',')[0]+", "+entry.split(',')[1] 
    return entry

def cleanNameEntry(entry):
  entry_list = []
  if ';' not in entry:
    entry = cleanEntry(entry)
    return entry
  else:
    for e in entry.split(';'):
      entry_list.append(cleanEntry(e))
    return entry_list
```

```python
def formatDF(raw_email):
  format_check_string = "\n\nSent: \nTo:" ## Should come back to this as a way of determining which type of email we're looking at
  if raw_email.startswith(format_check_string):
    p = "(?<=Subject:)(.*)(?=Re:)[^\n]+"
    p_search = re.search(p, raw_email, re.DOTALL)
    if p_search:
      from_line = p_search.group().strip().split('\n')[0] 
      sent_line = p_search.group().strip().split('\n')[2]
      to_line = p_search.group().strip().split('\n')[3]
      cc_line = p_search.group().strip().split('\n')[5]
      subject_line = p_search.group().strip().split('\n')[6]
      body_re = "(?<=Re:)(.*)"
      body_search = re.search(body_re, raw_email, re.DOTALL)
      if body_search:
        body = body_search.group().replace(subject_line, '').strip()
      else:
        body = None 
    else:
      from_line = None
      sent_line = None
      to_line = None
      cc_line = None
      subject_line = None
      body = None 
  else:
    from_line = extractFrom(raw_email)
    sent_line = extractDate(raw_email)
    to_line = extractTo(raw_email)
    cc_line = extractCc(raw_email)
    subject_line = extractSubject(raw_email)
    body = extractBody(raw_email, subject_line)

  return from_line, sent_line, to_line, cc_line, subject_line, body
```

```python
def checkNameOrEmail(entry):
  name_re = r"([A-Z][a-z]+, [A-Z][a-z]+)"
  name_search = re.search(name_re, entry)

  email_re = r"([a-zA-Z0-9._-]+@[a-zA-Z0-9._-]+\.[a-zA-Z0-9_-]+)"
  email_search = re.search(email_re, entry)
  
  if name_search or email_search:
    return True
  else:
    return False
```

```python
def findConnectedGraph(edgelist_df):
  G = nx.from_pandas_edgelist(edgelist_df, 'From', 'To', edge_attr=True)
  components = list(G.subgraph(c) for c in nx.connected_components(G))
  connected_graph = components[0]

  pos = nx.drawing.layout.spring_layout(connected_graph)
  nx.set_node_attributes(connected_graph, pos, 'pos')

  return connected_graph
```

```python
def plotNetwork(graph, metric):
  import plotly.graph_objects as go

  edge_x = []
  edge_y = []
  for edge in graph.edges():
      x0, y0 = graph.nodes[edge[0]]['pos']
      x1, y1 = graph.nodes[edge[1]]['pos']
      edge_x.append(x0)
      edge_x.append(x1)
      edge_x.append(None)
      edge_y.append(y0)
      edge_y.append(y1)
      edge_y.append(None)

  edge_trace = go.Scatter(
      x=edge_x, y=edge_y,
      line=dict(width=0.5, color='#888'),
      hoverinfo='none',
      mode='lines')

  node_x = []
  node_y = []
  for node in graph.nodes():
      x, y = graph.nodes[node]['pos']
      node_x.append(x)
      node_y.append(y)

  node_trace = go.Scatter(
      x=node_x, y=node_y,
      mode='markers',
      hoverinfo='text',
      marker=dict(
          showscale=True,
          # colorscale options
          #'Greys' | 'YlGnBu' | 'Greens' | 'YlOrRd' | 'Bluered' | 'RdBu' |
          #'Reds' | 'Blues' | 'Picnic' | 'Rainbow' | 'Portland' | 'Jet' |
          #'Hot' | 'Blackbody' | 'Earth' | 'Electric' | 'Viridis' |
          colorscale='Electric',
          reversescale=True,
          color=[],
          size=10,
          colorbar=dict(
              thickness=15,
              title='Node Connections',
              xanchor='left',
              titleside='right'
          ),
          line_width=2))
  if metric == 'degree':
  ## Degree Measure
    node_adjacencies = []
    node_text = []
    node_custom = []
    for node, adjacencies in enumerate(graph.adjacency()):
        node_adjacencies.append(len(adjacencies[1]))
        node_text.append(f'{str(adjacencies[0])} has '+str(len(adjacencies[1]))+' connection(s)')
    print(node_adjacencies)

    node_trace.marker.color = node_adjacencies
    node_trace.text = node_text

  elif metric == 'betweenness':
  ## Betweenness Measure
    node_scores = []
    node_text = []
    node_custom = []
    for node, score in nx.betweenness_centrality(graph).items():
        node_scores.append(score)
        node_text.append(f'{str(node)} has '+str(score)+' betweenness')
    print(node_scores)

    node_trace.marker.color = node_scores
    node_trace.text = node_text

  elif metric == 'pagerank':
  ## Pagerank Measure
    node_scores = []
    node_text = []
    node_custom = []
    for node, score in nx.pagerank(graph).items():
        node_scores.append(score)
        node_text.append(f'{str(node)} has '+str(score)+' pagerank')
    print(node_scores)

    node_trace.marker.color = node_scores
    node_trace.text = node_text

  ## @TODO Build out type metric -- connect to betweenness?
  # node_type = []
  # node_text = []
  # for edge in graph.edges():
  #     node_type.append(graph.nodes[edge[0]]['Source'])
  #     node_text.append(f'{str(graph.nodes[edge[0]])}')

  # node_trace.marker.color = node_type
  # node_trace.text = node_text

  fig = go.Figure(data=[edge_trace, node_trace],
              layout=go.Layout(
                  showlegend=False,
                  hovermode='closest',
                  margin=dict(b=20,l=5,r=5,t=40),
                  xaxis=dict(showgrid=False, zeroline=False, showticklabels=False),
                  yaxis=dict(showgrid=False, zeroline=False, showticklabels=False))
                  )
  fig.show()
```

```python
def findBetweennessDelta(sub_graph, connected_graph):
  zero_betweenness = []
  connected_betweenness = []

  for node, score in nx.betweenness_centrality(sub_graph).items():
    if score == 0:
      zero_betweenness.append(node)
  for node, score in nx.betweenness_centrality(connected_graph).items():
    if (node in zero_betweenness) and (score != 0):
      connected_betweenness.append((node, score))

  return connected_betweenness
```

```python
def getSpacyDF(name, org_df):
  corpus = org_df.loc[(org_df['From'] == name) |(org_df['To'] == name)]['Body']
  cv = CountVectorizer(max_df=0.95, min_df=2, stop_words='english', ngram_range=(1,1)) 
  dtm_body = cv.fit_transform(corpus)
  features = cv.get_feature_names_out()

  word_dict = {}
  lemmatized = {}
  for index, freq in enumerate(np.sum(dtm_body.toarray(), axis=0)):
    if not features[index].isnumeric():
      word_dict[features[index]] = freq

  for key in word_dict:
    doc = nlp(key)
    for d in doc:
      lemma = d.lemma_
      if (not bool(re.search(r'\d', lemma))) and (len(lemma) > 3) and (d.text not in names) and (not d.is_oov):
        if lemma in lemmatized:
          lemmatized[d] = lemmatized[lemma] + word_dict[key]
        else:
          lemmatized[d] = word_dict[key]

  for key in lemmatized:
    lemmatized[key] = (lemmatized[key]/dtm_body.sum())

  lemmatized = dict(sorted(lemmatized.items(), key=lambda item: item[1], reverse=True))

  spacy_df = pd.DataFrame.from_dict(lemmatized, orient='index').reset_index().rename(columns={'index':'word'})[:20]
  spacy_df['vector'] = spacy_df['word'].apply(lambda x: x.vector)

  return spacy_df
```

```python
def generateSimilarityMatrix(name_dfA, name_dfB):
  corr = cdist(np.array(name_dfA['vector'].values.tolist()), np.array(name_dfB['vector'].values.tolist()), metric='cosine')
  return corr
```

```python
def getWordFreqStats(name, org_df):
  corpus = org_df.loc[(org_df['From'] == name) |(org_df['To'] == name)]['Body']
  cv = CountVectorizer(max_df=0.95, min_df=2, stop_words='english', ngram_range=(1,1)) 
  dtm_body = cv.fit_transform(corpus)
  features = cv.get_feature_names_out()

  word_dict = {}
  lemmatized = {}
  for index, freq in enumerate(np.sum(dtm_body.toarray(), axis=0)):
    if not features[index].isnumeric():
      word_dict[features[index]] = freq

  for key in word_dict:
    doc = nlp(key)
    for sentence in doc.sentences:
      for word in sentence.words:
        lemma = word.lemma
        if not bool(re.search(r'\d', lemma)):
          if lemma in lemmatized:
            lemmatized[lemma] = lemmatized[lemma] + word_dict[key]
          else:
            lemmatized[lemma] = word_dict[key]

  for key in lemmatized:
    lemmatized[key] = (lemmatized[key]/dtm_body.sum())*(10**6) ## frequency per million words

  lemmatized = dict(sorted(lemmatized.items(), key=lambda item: item[1], reverse=True))

  return lemmatized
```

```python
def createComparison(degree, connector):
  compare_dict = {}
  for key in degree:
    for c_key in connector:
      compare_list = []
      for k in connector[c_key]:
        if ((k not in degree[key]) and (connector[c_key][k] > 6000)): ## arbitrary value
          compare_list.append((k, connector[c_key][k]))
      compare_dict[f'{key}-{c_key}'] = (len(compare_list), compare_list)
  return compare_dict
```
